(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

g = 9.8 #m/s^2
ell = 0.5 #m
#simple pendulum ODE's and conditions
def deriv(theta_current,t):  # return derivatives of the array theta
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

t = np.arange(0.,50.+0.01, 0.01)

omega0 = 0

th0 = np.pi/4
for i in range(5):
    theta_init = np.array([float(i*th0), omega0])
    theta = odeint(deriv,theta_init,t)
    plt.plot(t,theta[:,0], label = "%s * pi" %(i))
    plt.legend()
    plt.grid()
```

As the initial angle increases from  pi/4 to pi, the amplitude of the oscillator increases, as one would expect. At pi, the oscillator is in unstable equilibrium, and perturbs into a revolution. The oscillator comes to rest and repeats this cycle, remaining at rest for successively shorter time intervals. At least, this is what I speculate happens.

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dydx=x+2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.


```python
def EulerFreefall():
    tmax = 10 # maximum time
    dt = 0.1 # time step
    a = -9.81 #g in m/s^2
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i]
        
    return t,v,y
```
