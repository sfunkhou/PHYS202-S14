## Question 1

For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE Tour.

(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
#from IPython.html.widgets import interact, interactive

g = 9.8 #m/s^2
ell = 0.5 #m
#simple pendulum ODE's and conditions
def deriv(theta_current,t):  # return derivatives of the array theta
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

#def simplePendulum()
t = np.arange(0.,20.+0.01, 0.01)

omega0 = 0

theta = np.pi/4
for i in range(1,5):
    theta_init = np.array([float(i*theta), omega0])
    plt.plot(t,odeint(deriv,theta_init,t)[:,0], label = "%s * %s/4" % (i, r'$pi$'))
    plt.legend()
    #plt.grid()
    plt.xlabel(r'$t (seconds)$',fontsize=20)
    plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.grid()
```

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
xmax = 10
dx = 0.25
d2ydx = 1

nsteps = int(xmax/dx)

x = np.zeros(nsteps)
dydx = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0] = 0
y[0] = 0


for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    dydx[i+1] = dydx[i] + dx*d2ydx
    y[i+1] = y[i] + dx * dydx[i]
    if x[i] == 1:
        print "y(1) = ", y[i]
print y
#print y[2]
```
