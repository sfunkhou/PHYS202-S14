# Differentiation

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.,10.01,0.1)
y = x ** 5

dy = np.diff(y)

print len(y), len(dy)
dx = np.diff(x)
dydx = (dy/dx)
print len(dydx)
```

```python
#can't plot just yet, since x, and thus dx, contains zero as first entry
#plt.plot(x,dydx)

dydx1 = np.zeros(y.shape,float)

dydx1[0:-1] = np.diff(y)/np.diff(x)
dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
```

```python
#center-of-bin method:
dydx2 = np.zeros(y.shape,float)

dydx2[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2])

dydx2[0] = (y[1] -y[0])/(x[-1] - x[-2])

dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
```

```python
dydxExact = 5*x**4

plt.plot(x,dydx1,label='forward difference')
plt.plot(x,dydx2, label = 'center difference')
plt.plot(x,dydxExact, label= 'exact')
plt.legend(loc='upper left')
plt.show()
```

```python
percentError1 = 100*abs( dydx1[1:] - dydxExact[1:] )/dydxExact[1:]
percentError2 = 100*abs( dydx2[1:] - dydxExact[1:] )/dydxExact[1:]

#semilog y-axis to see variation better
plt.semilogy(x[1:], percentError1, label='forward difference')
plt.semilogy(x[1:], percentError2, label='center difference')
plt.ylabel("percent error with dydxExact")
plt.legend(loc="upper right")
plt.show()
```
