# Avogadro's Number: Image Segmentation and Bead Identification

This notebook furnishes several functions that will identify blobs immersed in a homogenous liquid based on their luminance (depth), and their aggregation into discrete beads (based on the given condition of a maximum bead radius, P) as depicted in a sequence of jpeg images.

**Components:**

1) Image Segmemtation (Blob/Pixel Identification)
   
       i) BlobFinder 
           
           Uses modified versions of "monochrome" and "FillRec" in "CountingStars" 
           notebook to recolor contiguous beads based on luminance

2) Particle (Bead) Identification
   
       i) add(i,j) 
       
           add pixel (i,j) to the blob, based on BlobFinder
           
       ii) mass() 
   
           return number of pixels in bead
           
       iii) distanceTo(blob) 
       
           returns distance between centers of masses of this bead and blob
           
       iv) centerOfMass() 
       
           return tuple of (x,y) values for this blob's center of mass
