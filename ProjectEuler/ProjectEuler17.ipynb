# Project Euler 17

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage

```python
import numpy as np
#numbers = np.zeros(1000)

number = ["one", "two", "three", "four", "five", "six", "seven", "eight",
          "nine", "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen",
          "sixteen", "seventeen", "eighteen", "nineteen,"]
tens = ["twenty", "thirty", "fourty", "fifty", "sixty", "seventy",
        "eighty", "ninety"]

for u in range(8):
    number.append(tens[u])
    for i in range(9):
        number.append(tens[u] + number[i])#no spaces
#now have up to 29

for u in range(10):
    number.append(number[u]+"hundred")
    for i in range(99):
        number.append(number[u] + "hundredand" + number[i])
    

count = 0
for i in range(1000):
    count += len(number[i])

print count
#print number
```
