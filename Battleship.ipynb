This is a primitive Battleship game. There is one battleship that spans five squares somewhere on a 12 x 12 grid. The first square the battleship contains is selected randomly, as well as the second square, whose position influences the direction by which the remaining three squares are appended to the body of the ship.

The user selects how many guesses (turns) to take. For each guess, the user chooses a column number and row number. Hits and misses are displayed as red and white squares. At this time, the ship only occupies contiguous vertical or horizontal squares.

```python
from random import randint
from ipythonblocks import BlockGrid

print "Let's play Battleship!"
#print "How many columns would you like to play with? (Note, the battleship is 5 squares long).

board = BlockGrid(12, 12, fill = (95, 158, 160))

board.show()

turns = int(raw_input("How many turns would you like?"))

def random_row(lower, upper):
    return randint(lower, upper)

def random_col(lower, upper):
    return randint(lower, upper)

ship_row = [0,0,0,0,0]
ship_col = [0,0,0,0,0]

#for now, making it so that first square is not on a border, to simplify arranging the ship
ship_row[0] = random_row(0, board.width - 2)
ship_col[0] = random_col(0, board.height - 2)

#randomly selecting second contiguous square (not diagonal as of yet)
i = 0
while (i == 0):
    ship_row[1] = random_row(ship_row[0] - 1, ship_row[0] + 1)
    ship_col[1] = random_col(ship_col[0] - 1, ship_col[0] + 1)
    
    if ship_col[1] != ship_col[0] or ship_row[1] != ship_row[0]:
        i = 1
    #no diagonal positions:
    if ship_col[1] != ship_col[0] and ship_row[1] != ship_row[0]:
        i = 0

for i in range(2,5):
    #if horizontal...           and ...accruing to the right, and last point not on border
    if ship_row[i - 2] == ship_row[i - 1] and ship_col[i - 1] > ship_col[i - 2] and ship_col[i - 1] < 11:
        ship_col[i] = ship_col[i - 1] + 1
        ship_row[i] = ship_row[1]
    #if horizontal and last point was on border:
    if ship_row[i - 2] == ship_row[i - 1] and ship_col[i - 1] == 11:
        ship_col[i] == ship_col[0] - 1
        ship_row[i] = ship_row[1]
    #if horizontal and accruing left and last point not on border
    if ship_row[i - 2] == ship_row[i - 1] and ship_col[i - 1] < ship_col[i - 2] and ship_col[i - 1] > 0:
        ship_col[i] = ship_col[i - 1] - 1
        ship_row[i] = ship_row[1]
    #if horizontal and last point on border (0)
    if ship_row[i - 2] == ship_row[i - 1] and ship_col[i - 1] == 0:
        ship_col[i] = ship_col[0] + 1
        ship_row[i] = ship_row[1]
        
    #if vertical...           and ...accruing down, and last point not on border
    if ship_col[i - 2] == ship_col[i - 1] and ship_row[i - 1] > ship_row[i - 2] and ship_row[i - 1] < 11:
        ship_row[i] = ship_row[i - 1] + 1
        ship_col[i] = ship_col[1]
    #if vertical and last point was on border:
    if ship_col[i - 2] == ship_col[i - 1] and ship_row[i - 1] == 11:
        ship_row[i] == ship_row[0] - 1
        ship_col[i] = ship_col[1]
    #if vertical and accruing up and last point not on border
    if ship_col[i - 2] == ship_col[i - 1] and ship_row[i - 1] < ship_row[i - 2] and ship_row[i - 1] > 0:
        ship_row[i] = ship_row[i - 1] - 1
        ship_col[i] = ship_col[1]
    #if horizontal and last point on border (0)
    if ship_col[i - 2] == ship_col[i - 1] and ship_row[i - 1] == 0:
        ship_row[i] = ship_row[0] + 1
        ship_col[i] = ship_col[1]
    
for i in range(5):
    print ship_row[i], ship_col[i]

# Everything from here on should go in your for loop!
# Be sure to indent four spaces!
for turn in range(turns):
    
    hits = 0
    guess_row = int(raw_input("Guess Row:"))
    guess_col = int(raw_input("Guess Col:"))

    if guess_row in ship_row and guess_col in ship_col:
        print "Hit!"
        board[guess_row, guess_col] = (255, 0 , 0)
        hits += 1
        board.show()
    else:
        if (guess_row < 0 or guess_row > 11) or (guess_col < 0 or guess_col > 11):
            print "Oops, that's not even in the ocean."
        elif board[guess_row, guess_col] == (255, 255, 255):
            print "You guessed that one already."
        else:
            print "You missed my battleship!"
            board[guess_row, guess_col] = (255, 255, 255)
            board.show()
    if hits == 5 and turn <= turns:
        print "You sank the battleship!"
    if turn == turns - 1:
        print "Game Over"
        for i in range(5):
            if board[ship_row[i],ship_col[i]] != (255, 0, 0):
                board[ship_row[i],ship_col[i]] = (0, 0, 0)
        board.show()
        
    # Print (turn + 1) here!
    print "Turn", turn + 1
    
    
    
```
